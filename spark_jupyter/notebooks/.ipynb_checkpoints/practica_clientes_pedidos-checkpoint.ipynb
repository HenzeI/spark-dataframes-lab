# Práctica · Spark DataFrames · Clientes y pedidos

Completa las celdas indicadas. Trabaja sobre el entorno dockerizado incluido en `spark_jupyter/`.

```python
from iniciar_spark import get_spark
from pyspark.sql import functions as F

spark = get_spark("PracticaClientesPedidos")
print("SparkSession creada")
print("Versión de Spark:", spark.version)
```

```python
from pathlib import Path

data_dir = Path("/opt/spark-apps/datos")
clientes_path = data_dir / "clientes.csv"
pedidos_path = data_dir / "pedidos.csv"

print("Ruta clientes:", clientes_path)
print("Ruta pedidos:", pedidos_path)
```

## 1. Lee ambos CSV y muestra su esquema y algunas filas.

```python
# Leer clientes
clientes_df = (
    spark.read
    .option("header", True)
    .option("sep", ";")
    .csv(str(clientes_path))
)
print("=== Esquema clientes ===")
clientes_df.printSchema()
clientes_df.show(5, truncate=False)

# Leer pedidos
pedidos_df = (
    spark.read
    .option("header", True)
    .option("sep", ";")
    .csv(str(pedidos_path))
)
print("=== Esquema pedidos ===")
pedidos_df.printSchema()
pedidos_df.show(5, truncate=False)
```

## 2. Limpia el DataFrame de clientes: trim en ciudad y dropDuplicates.

```python
clientes_clean = (
    clientes_df
    .withColumn("ciudad", F.trim(F.col("ciudad")))
    .withColumn("nombre", F.trim(F.col("nombre")))
    .withColumn("segmento", F.trim(F.col("segmento")))
    .dropDuplicates()
    .dropna(subset=["id_cliente"])
    .withColumn("id_cliente", F.col("id_cliente").cast("integer"))
)

print(f"Filas antes de limpiar: {clientes_df.count()}")
print(f"Filas después de limpiar: {clientes_clean.count()}")
clientes_clean.show(truncate=False)
```

## 3. Convierte tipos en pedidos y crea la columna `importe`.

```python
pedidos_typed = (
    pedidos_df
    .withColumn("id_pedido",       F.col("id_pedido").cast("integer"))
    .withColumn("id_cliente",      F.col("id_cliente").cast("integer"))
    .withColumn("fecha",           F.col("fecha").cast("date"))
    .withColumn("cantidad",        F.col("cantidad").cast("double"))
    .withColumn("precio_unitario", F.col("precio_unitario").cast("double"))
    .withColumn("importe",         F.round(F.col("cantidad") * F.col("precio_unitario"), 2))
)

print("=== Esquema pedidos con tipos corregidos ===")
pedidos_typed.printSchema()

print(f"\nPedidos con cantidad nula:  {pedidos_typed.filter(F.col('cantidad').isNull()).count()}")
print(f"Pedidos con producto nulo:  {pedidos_typed.filter(F.col('producto').isNull()).count()}")
print(f"Pedidos con importe nulo:   {pedidos_typed.filter(F.col('importe').isNull()).count()}")

pedidos_typed.show(10)
```

## 4. Haz un join entre clientes y pedidos.

```python
# Inner join: solo clientes que tienen pedidos y pedidos con cliente conocido
df_join = clientes_clean.join(pedidos_typed, on="id_cliente", how="inner")

print(f"Pedidos totales:           {pedidos_typed.count()}")
print(f"Registros tras inner join: {df_join.count()}")

# Pedidos perdidos: id_cliente no existe en clientes
print("\nPedidos perdidos (id_cliente sin match en clientes):")
pedidos_typed.join(clientes_clean, on="id_cliente", how="left_anti") \
             .select("id_pedido", "id_cliente", "producto") \
             .orderBy("id_cliente") \
             .show()

df_join.show(10)
```

## 5. Filtra ventas Premium con importe >= 100.

```python
df_premium = df_join.filter(
    (F.col("segmento") == "Premium") & (F.col("importe") >= 100)
)

print(f"Ventas Premium con importe >= 100: {df_premium.count()}")
df_premium.select("id_pedido", "nombre", "ciudad", "segmento", "producto", "importe") \
          .orderBy(F.col("importe").desc()) \
          .show(20)
```

## 6. Clasifica pedidos con `when` en Alto / Medio / Bajo.

```python
df_clasificado = df_join.withColumn(
    "categoria_importe",
    F.when(F.col("importe") >= 500, "Alto")
     .when(F.col("importe") >= 100, "Medio")
     .when(F.col("importe").isNotNull(), "Bajo")
     .otherwise("Sin importe")
)

df_clasificado.select("id_pedido", "nombre", "producto", "importe", "categoria_importe") \
              .orderBy("id_pedido") \
              .show(20)

df_clasificado.groupBy("categoria_importe").count().orderBy("categoria_importe").show()
```

## 7. Calcula agregaciones por ciudad y segmento.

```python
agg_df = (
    df_clasificado
    .groupBy("ciudad", "segmento")
    .agg(
        F.count("id_pedido").alias("num_pedidos"),
        F.round(F.sum("importe"), 2).alias("ingresos_totales"),
        F.round(F.avg("importe"), 2).alias("importe_medio")
    )
    .orderBy(F.col("ingresos_totales").desc())
)

print(f"Combinaciones ciudad/segmento: {agg_df.count()}")
agg_df.show(30)
```

## 8. Crea una vista temporal y haz una consulta SQL.

```python
df_clasificado.createOrReplaceTempView("pedidos_clientes")

resultado_sql = spark.sql("""
    SELECT
        ciudad,
        segmento,
        producto,
        COUNT(*)                    AS num_pedidos,
        ROUND(SUM(importe),  2)     AS ingresos_totales,
        ROUND(AVG(importe),  2)     AS importe_medio,
        ROUND(MAX(importe),  2)     AS importe_max
    FROM pedidos_clientes
    WHERE importe IS NOT NULL
    GROUP BY ciudad, segmento, producto
    ORDER BY ingresos_totales DESC
    LIMIT 15
""")

print("=== Top 15 combinaciones ciudad/segmento/producto por ingresos ===")
resultado_sql.show(truncate=False)
```

## 9. Usa `sample()` y `randomSplit()`.

```python
total = df_clasificado.count()

# --- sample() ---
muestra = df_clasificado.sample(withReplacement=False, fraction=0.2, seed=42)
print(f"Total registros:       {total}")
print(f"Muestra (fraction=0.2): {muestra.count()} filas aprox.")
muestra.select("id_pedido", "nombre", "producto", "importe").show(5)

# --- randomSplit() ---
train, test = df_clasificado.randomSplit([0.8, 0.2], seed=42)
print(f"\nrandomSplit  →  train: {train.count()} filas | test: {test.count()} filas")
print(f"Proporción real: train={train.count()/total:.1%}  test={test.count()/total:.1%}")
```

## 10. Guarda el resultado en Parquet en `/opt/spark-apps/salida/resultado_parquet` y léelo de nuevo.

```python
output_path = "/opt/spark-apps/salida/resultado_parquet"

# Guardar en Parquet (sobreescribir si ya existe)
agg_df.write.mode("overwrite").parquet(output_path)
print(f"Resultado guardado en: {output_path}")

# Leer de nuevo
df_parquet = spark.read.parquet(output_path)
print("\n=== Esquema leído desde Parquet ===")
df_parquet.printSchema()
df_parquet.show()
```

## 11. Responde brevemente en Markdown:
- ¿Qué ventaja tiene usar join?
- ¿Qué diferencia hay entre sample y randomSplit?
- ¿Qué pedido se pierde en el inner join y por qué?

### Respuestas

**¿Qué ventaja tiene usar join?**  
El `join` permite combinar en un solo DataFrame la información de dos fuentes distintas (clientes y pedidos) usando una clave común (`id_cliente`). Esto evita tener que iterar manualmente sobre los datos y aprovecha la ejecución distribuida de Spark: el plan de ejecución optimiza la operación (broadcast join si un lado es pequeño, sort-merge join en otros casos) sin mover datos innecesariamente entre nodos.

**¿Qué diferencia hay entre `sample` y `randomSplit`?**  
- `sample(fraction=f)` devuelve **un único** subconjunto aleatorio de aproximadamente `f × N` filas. Es útil para exploración rápida o pruebas con un porcentaje del dataset.  
- `randomSplit([p1, p2, ...])` divide el DataFrame en **múltiples** subconjuntos disjuntos cuyas proporciones suman 1. Se usa habitualmente para separar datos en conjuntos de entrenamiento y prueba (train/test split) garantizando que cada fila aparece en exactamente uno de los subconjuntos.

**¿Qué pedido se pierde en el inner join y por qué?**  
Se pierden los pedidos cuyos `id_cliente` no existen en el DataFrame de clientes: en los datos de ejemplo los pedidos con `id_cliente` 41, 42, 45, 46, 47 y 48 no tienen cliente correspondiente (el CSV de clientes solo llega hasta el `id_cliente` 40). Al usar `inner join` solo se conservan las filas con match en ambos lados, por lo que esos pedidos quedan fuera. Para conservarlos habría que usar un `left join` (desde pedidos) o un `full outer join`.
